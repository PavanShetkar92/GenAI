Task 5 create product in to file (User input)

```python
#Get the number of products count from user
n = int(input("Enter the number of products you want to add: "))

#Get the products and their price from user
with open('products.txt','w') as f:
    for i in range(n):
        product = input("Enter product name: ")
        price = input("Enter the price: ")
        f.write(f"{product}|{price}\n")

#Read and display the products
with open('products.txt','r') as f:
    products = f.readlines()
    for product in products:
        print(product.strip())
```

Task 7 | Mini Project

```python
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

#get discount percentage from user
discount_percent = float(input("Enter the discount percent: "))

#calculate the discount price from given discount percent and write all these in discount_report.txt
with open('discount_report.txt','w') as f:
    f.write("Product|Original_Price|Discount_price\n")
    f.write("-------------------------------------\n")
    for product,orginal_price in prices.items():
        discount_price = orginal_price-(orginal_price*discount_percent/100) 
        f.write(f"{product}|{orginal_price}|{discount_price}\n")   

```

```python
#Read the discount_report.txt
with open('discount_report.txt','r') as f:
    final_list = f.readlines()
    for lines in final_list:
        print(lines.strip())

    
```
